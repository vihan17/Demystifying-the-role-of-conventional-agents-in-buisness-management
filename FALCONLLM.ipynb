```python
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer, pipeline
import torch

model = "tiiuae/falcon-7b-instruct" #tiiuae/falcon-40b-instruct

tokenizer = AutoTokenizer.from_pretrained(model)

pipeline = pipeline(
    "text-generation", #task
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=200,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
```

```python
llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs = {'temperature':0})
```

```python
from langchain import PromptTemplate,  LLMChain
```

```python
template = """
You are an intelligent chatbot. Help the following question with brilliant answers.
Question: {question}
Answer:"""
prompt = PromptTemplate(template=template, input_variables=["question"])
```

```python
llm_chain = LLMChain(prompt=prompt, llm=llm)
```

```python
question = "Why is your customer service so bad. Just tell me what are your working hours."

print(llm_chain.run(question))
```

```python
question = "I am angry. Address my feelings and then answer this question: Why is your customer service so bad. Just tell me what are your working hours."

print(llm_chain.run(question))
```
